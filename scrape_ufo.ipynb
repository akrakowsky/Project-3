## UFO Sightings

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.ufo_db
collection = db.ndxevent
```

```python
# URL of page to be scraped
url = 'http://www.nuforc.org/webreports/ndxevent.html'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
response.text
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

url = 'http://www.nuforc.org/webreports/ndxevent.html'
browser.visit(url)
```

```python
data = browser.find_by_css("td a")
```

```python
ufo_links = [x["href"] for x in data]
```

```python
browser.quit()
```

```python
df_list = []
for index,i in enumerate(ufo_links):
    df = pd.read_html(i)[0]
    df_list.append(df)
    print(index)
    time.sleep(1)
```

```python
df_list
```

```python
ufo_sightings = pd.concat(df_list)
ufo_sightings
```

```python
ufo_sightings['Location'] = ufo_sightings['City'] + ", " + ufo_sightings['State']
ufo_sightings
```

```python
ufo_sightings = ufo_sightings.reset_index(drop=True)
ufo_sightings
```

```python
ufo_sightings = ufo_sightings.dropna(how="all", subset=["Location"])
ufo_sightings
```

```python
cities = pd.read_csv("us_cities.csv")
cities
```

```python
ufo_locations = ufo_sightings.merge(cities, how="inner", left_on=["State", "City"], right_on=["STATE_CODE", "CITY"])
ufo_locations
```

```python
ufo_locations.to_csv("ufo_sightings_locations.csv")
```
